```python
Num_Coulmns = x.select_dtypes(include = ["float64"]).columns
for col in Num_Coulmns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=x[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
for col in Num_Coulmns:
    Q1 = x[col].quantile(0.25)
    Q3 = x[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = x[(x[col] < lower_bound) | (x[col] > upper_bound)][col]
    print(f"{col}: {len(outliers)} outliers")
```

```python
Num_Coulmns = x.select_dtypes(include = ["float64","int64"]).columns.tolist()
print("Numeric columns:")
print("==="*8)
for i , col in enumerate(Num_Coulmns):
    if x[col].isnull().sum() > 0:
        print(f"{i + 1}-{col} ===> {x[col].isnull().sum()} Null Values")
```

```python
median_fill = ['blood_pressure', 'heart_rate', 'insulin', 'daily_steps', 'income']
mean_fill = ['gene_marker_flag']
for col in mean_fill:
    if x[col].isnull().sum() > 0:
        x[col] = x[col].fillna(x[col].mean())

for col in median_fill:
    if x[col].isnull().sum() > 0:
       x[col] = x[col].fillna(x[col].median())

```

```python
Num_Coulmns = x.select_dtypes(include = ["float64","int64"]).columns.tolist()
print("Numeric columns:")
print("==="*8)
for i , col in enumerate(Num_Coulmns):
    print(f"{i + 1}-{col} ===> {x[col].isnull().sum()} Null Values")
```

```python
Categori_Coulmns = x.select_dtypes(include = ["object"]).columns.tolist()
print("Categori_Coulmns:")
print("==="*8)
Total = x.shape[0]   # ===> Total Rowes
for i, col in enumerate(Categori_Coulmns):
    missing_count = x[col].isnull().sum()
    if missing_count > 0:
        missing_percent = (missing_count / Total) * 100
        print(f"{i + 1}-{col} ===> {missing_count} Null Values and {missing_percent:.2f}%")
```

```python
cols = ['alcohol_consumption', 'exercise_type', 'caffeine_intake']

for col in cols:
    x[col] = x[col].fillna('Unknown')
# I Fill With "Unknown" Word Because Avoid Repetation And Because The Percentage The Null Values In This Coulmns > 20%.
```

```python
categorical_cols = x.select_dtypes(include = ["object"]).columns.tolist()

drop = []

for i , col in enumerate(categorical_cols):
    top_freq = x[col].value_counts(normalize=True).iloc[0]  
    print(f"{i + 1}- {col}: {top_freq:.2%} dominant class")
    print("==="*49)
    if top_freq >= 0.8:
        drop.append(col)
print("Columns to drop (dominant class >= 80%):")
print(drop)
print("==="*49)
```
